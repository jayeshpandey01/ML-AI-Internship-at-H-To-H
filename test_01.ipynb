```python
import yfinance as yf
import pandas as pd

# Parameters
tickers = ['RELIANCE.NS', 'TCS.NS', 'INFY.NS']
period = '5d'         # Use '5d' instead of '6mo' for 5-minute interval support
interval = '5m'

# Download data
raw_data = {}
for ticker in tickers:
    raw_data[ticker] = yf.download(ticker, period=period, interval=interval)

# Store individual dataframes
reliance_df = raw_data['RELIANCE.NS']
tcs_df = raw_data['TCS.NS']
infy_df = raw_data['INFY.NS']
```
